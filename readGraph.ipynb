```python
import numpy as np

mode  = 'normal'
datas = []
sums  = 0
for xyear in [2000,2005,2010,2015]:
    cyear = xyear + 5
    file  = f'Result/graph_{mode}/Graph_pred_{xyear}_{cyear}.txt'
    data = np.loadtxt(file, dtype=int, delimiter=',')
    datas.append(data)
    sums = sums + data

sums  
```
